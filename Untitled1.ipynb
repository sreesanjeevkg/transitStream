```python
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files.sort()
files
```

```python
P = np.load(files[0])
S = np.load(files[1])
T = np.load(files[2])
date = np.load(files[3])
lat = np.load(files[4])
levels = np.load(files[5])
lon = np.load(files[6])
```

```python
da_salinity = xr.DataArray(S, dims=['level', 'date'],
                           coords={'level': levels,
                                   'date': date},)
da_salinity
```

```python
from pathlib import Path
from datetime import datetime, timedelta
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
WATER = "darkgray"
LAND = "dimgray"
fig = plt.figure(figsize=(16, 16), facecolor="k")
ax = fig.add_axes([0, 0, 1, 1])
m = Basemap(projection="npstere", boundinglat=60, lon_0=315, resolution="l")
m.drawcoastlines(color="w")
m.drawparallels(np.arange(-0.0, 81.0, 20.0))
m.drawmeridians(np.arange(-10.0, 181.0, 20.0))
m.drawmapboundary(fill_color=WATER, color="k")
m.fillcontinents(color=LAND, lake_color=WATER)
```
