```python
from kafka import KafkaConsumer
from datetime import datetime
import time
```

```python
consumer = KafkaConsumer(bootstrap_servers='localhost:9092', key_deserializer = lambda s: s.decode(),
                            value_deserializer = lambda s: s.decode(), group_id = "basic-consumer",
                            auto_offset_reset = "earliest")
```

```python
consumer.subscription()
```

```python
consumer.subscribe("getting-started")
while(True):
    records = consumer.poll(1000)
    for topic_partition, messages in records.items():
        for message in messages:
            print(f"Key: {message.key}, Value: {message.value}")
            print("--------------------------------------")
    print("****************************************************************")
```

```python
consumer.close()
```
